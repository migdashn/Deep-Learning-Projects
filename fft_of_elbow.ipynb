```python
import numpy as np
import glob 
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
path = f"/content/gdrive/MyDrive/*.npy"
a = np.load("/content/gdrive/MyDrive/פרויקט גמר/VideoDataBase/ExercisesByExpert/processed/data/ARO.npy")
y= a[0:900,6,1]
y = (y - np.amin(y))/np.amax(y) 
y = y - np.amax(y)/2
x = np.arange(y.size)
print(x.size)
plt.plot(x,y)
plt.xlabel('Frames')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.show()
f = np.fft.fft(y)
N = f.size
T = 1/N
freq = np.fft.fftfreq(x.shape[-1])
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
#print(2.0/N * np.abs(f[:N//2]))
f[0] = 0
f[1] = 0
#plt.plot(freq,f.real,label='Real')
#plt.plot(freq,f.imag,label='Imaginary')
plt.plot(freq,np.abs(f))
plt.xlim([-0.05,0.05])
plt.xlabel('Freq (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
#plt.legend()
plt.show()
```
